3. Створіть масив розміром 9x9, заповніть його випадковими натуральними числами від 1 до 50 та знайдіть відсоток чисел, які діляться на 3 без остачі.

5. Створіть два одновимірних масиви розміром 7, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та виконайте на них поелементні операції додавання, віднімання та множення.

```python
a = np.random.randint(1, 6, size=7)
b = np.random.randint(1, 6, size=7)

print("a = ", a)
print("b = ", b)

print("a + b = ", a + b)
print("a - b = ", a - b)
print("a * b = ", a * b)
```

6. Створіть два вектори розміром 4, заповніть їх довільними дійсними числами та знайдіть їх скалярний добуток.

```python
a = np.random.randint(1, 11, size=4)
b = np.random.randint(1, 11, size=4)

dot_a_b = np.dot(a, b)

print("a = ", a)
print("b = ", b)
print("Скалярний добуток векторів дорівнює:", dot_a_b)
```

7. Створіть дві матриці розміром 2x3 та 3x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 15 та перемножте їх між собою.

```python
a = np.random.randint(1, 16, size=(2, 3))
b = np.random.randint(1, 16, size=(3, 4))

dot_a_b = np.dot(a, b)

print("a = ", a)
print("b = ", b)
print("Результат множення матриць:")
print(dot_a_b)
```

8. Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 15 та знайдіть її обернену матрицю.

```python
a = np.random.randint(1, 16, size=(2, 2))
print("Вихідна матриця:")
print(a)
if np.linalg.det(a) == 0:
    print("Матриця не має оберненої матриці.")
else:
    inv_a = np.linalg.inv(a)
    print("Обернена матриця:")
    print(inv_a)
```

9. Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 3 до 15 та транспонуйте її.

```python
a = np.random.randint(1, 16, size=(3, 3))
print(a)

a_transp = np.transpose(a)
print("Транспонована матриця:")
print(a_transp)
```

10. Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть суму добутків елементів кожного стовпця матриці на відповідний елемент вектору.

```python
matr = np.random.randint(1, 10, size=(3, 4))
vect = np.random.randint(1, 10, size=4)
print("Матриця:")
print(matr)
print("Вектор:")
print(vect)

print("Результат множення матриці на вектор:")
result = np.sum(matr * vect, axis=0)
print(result)
```

12. Створіть дві матриці розміром 8x8, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть добуток всіх елементів першої матриці.

13. Створіть дві матриці розміром 7x7, заповніть їх випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть різницю їх елементів.

14.	Створіть матрицю розміром 9x9, заповніть її випадковими дійсними числами в діапазоні від -10 до 10 та знайдіть суму елементів на головній діагоналі.

15.	Створіть матрицю розміром 9x9, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить суму елементів кожного стовпця матриці.

```python
a = np.random.uniform(-0.5, 0.5, size=(9, 9))
print("Перша матриця:")
print(a)
column_sums = a.sum(axis=0)
col_vec = column_sums.reshape((-1, 1))
print("Вектор-стовпчик з сумами елементів кожного стовпця:")
print(col_vec)
```
